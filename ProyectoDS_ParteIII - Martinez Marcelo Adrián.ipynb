```python
# 1 )Preparación de datos
X = Clinica[['NumReadmisiones', 'faltaHierro','desordenSangre', 'desordenPsicologico']]
y = Clinica['duracionEstancia']
```

```python
# ELIMINACIÓN DE CAMPOS PARA NUEVO DF
```

```python
ClinicaA= Clinica.drop(['FechaVisita', 'fechaAlta', 'idHospital'], axis=1)
print(ClinicaA)
```

```python
# ÁRBOLES DE REGRESIÓN
```

```python
# 
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import plot_tree
from sklearn.tree import export_graphviz
from sklearn.tree import export_text
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(
                                        ClinicaA.drop(columns = "duracionEstancia"),
                                        ClinicaA['duracionEstancia'],
                                        random_state = 42
                                    )
# Creación del modelo
# ------------------------------------------------------------------------------
modelo = DecisionTreeRegressor(
            max_depth         = 3,
            random_state      = 42
          )

# Entrenamiento del modelo
# ------------------------------------------------------------------------------
modelo.fit(X_train, y_train)
```

```python
# Estructura del árbol creado
# ------------------------------------------------------------------------------
fig, ax = plt.subplots(figsize=(12, 5))

print(f"Profundidad del árbol: {modelo.get_depth()}")
print(f"Número de nodos terminales: {modelo.get_n_leaves()}")

plot = plot_tree(
            decision_tree = modelo,
            feature_names = ClinicaA.drop(columns = "duracionEstancia").columns,
            class_names   = 'duracionEstancia',
            filled        = True,
            impurity      = False,
            fontsize      = 10,
            precision     = 2,
            ax            = ax
       )
```

```python
texto_modelo = export_text(
                    decision_tree = modelo,
                    feature_names = list(ClinicaA.drop(columns = "duracionEstancia").columns)
               )
print(texto_modelo)
```

```python
Siguiendo la rama más a la izquierda del árbol, puede verse que el modelo predice una duración de la 
estancia de 3 días aproximadamente para pacientes que tienen un número de readmisiones <= 1.5, tienen
un ddesorden Psicológico <= 0.5 y un desorden de sangre >= 0.5.
```

```python
importancia_predictores = pd.DataFrame(
                            {'predictor': ClinicaA.drop(columns = "duracionEstancia").columns,
                             'importancia': modelo.feature_importances_}
                            )
print("Importancia de los predictores en el modelo")
print("-------------------------------------------")
importancia_predictores.sort_values('importancia', ascending=False)
```

```python
Se puede rescatar de que la variable que tiene mayor peso a la hora de predecir la duración de la estancia
de los pacientes en la clínica es el "número de readmisiones".
Luego, con menor peso están las variables de "desorden Psicológico", "desorden de sangre" y pacientes que
presentan problemas de "hematocrito", siendo estos dos último los que menos importancia representan.
```

```python
# Error de test del modelo inicial
#-------------------------------------------------------------------------------
predicciones = modelo.predict(X = X_test)

rmse = mean_squared_error(
        y_true  = y_test,
        y_pred  = predicciones,
        squared = False
       )
print(f"El error (rmse) de test es: {rmse}")
```

```python
Esto indica que los valores que se predicen se desvían en 1.36 días de la duración de la estancia real de
los pacientes.
```

```python
print("Precisión del modelo: ")
print(modelo.score(X_train, y_train))
```

```python
Se puede establecer que el modelo tiene un nivel de precisión del 60%. 
```
